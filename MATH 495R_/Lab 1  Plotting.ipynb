```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Problem 1
x = np.linspace(np.pi*(-2),np.pi*2)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.arctan(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.xlim((-2)*np.pi,2*np.pi)
plt.show()
```

```python
#Problem 2
#Write a function to plot the curve f(x)= 1 x1 on the domain [-2,6].

x1 = np.linspace(-2,1)[:-1]
x2 = np.linspace(1,6)[1:]
y1 = 1/(x1-1)
y2 = 1/(x2-1)

plt.xlim(-2,6)
plt.ylim(-6,6)
plt.plot(x1,y1, "--",color= 'mediumorchid', linewidth = 4)
plt.plot(x2,y2, "--",color= 'mediumorchid', linewidth = 4)
plt.show()
```

```python
#Problem 3
#Create a plot that demonstrates your new found knowledge of plotting in Python. 
#It should involve at least three different functions, the axis should be labeled, the plot should have a legend, and should be labeled.

x = np.linspace(-10,10,50)
x1 = np.linspace(-10,0,50)[:-1]
x2 = np.linspace(0,10,50)[1:]
y = np.linspace(0,0,50)
plt.plot(x,y,"-", color="black")
plt.xlabel("The x axis")
plt.plot(y,x,"-", color="black")
plt.ylabel("The y axis")


plt.ylim(-5,5)
plt.plot(x2,1/x2,"-", linewidth=2, label="Quadrant 1")
plt.plot(x1,-1/x1,"o", markersize=5, label="Quadrant 2")
plt.plot(x1,1/x1,":", linewidth=2, label="Quadrant 3")
plt.plot(x2,-1/x2,"*", markersize=7, label="Quadrant 4")
plt.legend(loc="lower left") 

plt.show()
```
