```python
import numpy as np
import sympy as sy
import math as m
import matplotlib.pyplot as plt

#Problem 1-a)

n = sy.symbols("n")
y = sy.sin(n)
f = sy.lambdify(n,y)

plt.plot(np.arange(0,1000),f(np.arange(0,1000)), "o")
plt.show()

print("Conjecture: sin(n) has limit points [-1,1]")
```

```python
#Problem 1-c)
# a = a sequence, L = a suspected value of a limit point,and eps = epsilon
def subsequence_close_to_limit(a,m,L,eps):
  indices = []
  for n in range(0,m):
    if (abs(a[n]-L) < eps):
      indices.append(n)
  x = np.arange(0,m)
  sub_a = []
  for i in indices:
    sub_a.append(a[i])
  plt.plot(x,a, ".")
  plt.plot(indices, sub_a, ".", color = "r")
  plt.show()
  return indices
```

```python
#Problem 1-d)

L = 1/np.pi
eps = 0.01
a=[]
m=10001
for i in range(0,m):
  a.append(np.sin(i))

subsequence_close_to_limit(a,m,L,eps)

print("If L satisfies that -1 <= L <= 1, we can find such limit point L.")
print("The higher epsilon we set, the widther codomain we get. So, it is better to have small epsilon.")

```

```python
#Problem 2

L = -3/10
eps = 0.01
a=[]
m=10001
for i in range(0,10001):
  a.append(np.sin(np.pi*i/20))

subsequence_close_to_limit(a,m,L,eps)

print("The limit points of the new sequence has finitely many limit points because the new sequence is oscillating (21 limit points)")
```
