```python
import numpy as np
from matplotlib import pyplot as plt

# Problem 1

N = 50
a = []

for n in range(0,N):
  a.append(n/np.sqrt(n**2+1))

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Function 1")

plt.plot(a,".")
plt.show()

print("This function converges to 1")
```

```python
# Problem 2.a

N = 21
a = [(1-1/n)**n for n in range(1,N)]

plt.title("Problem 2.a")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xlim(-1, 20)
plt.ylim(0, 0.4)

plt.plot(a,"*", color = "g", label = "New")
plt.legend(loc = "upper left")
plt.show()
```

```python
# Problem 3.b

# My way of definding fib
def my_fib(n):
  f = []
  for i in np.arange(n+1):
    if(i < 2):
      f.append(i)
    else:
      f.append(f[i-1]+f[i-2])
  return f[n]

def r(n):
  if (n > 1):
    return my_fib(n)/my_fib(n-1)

N = 21
x = np.arange(2,N)
a = [r(n) for n in x]


plt.title("Problem 3.b")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xlim(0,20)

plt.plot(a, ".", color = "b", label = "r(x)")
plt.legend(loc = "lower right")
plt.show()
```
