```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

```python
#Problem 1
x = sy.symbols("x")
eps = x**2 - 1
f = sy.lambdify(x,eps)

iter = [x,f(x),f(f(x)),f(f(f(x))),f(f(f(f(x)))),f(f(f(f(f(x)))))]

lf = sy.lambdify(x,iter)

print("If x = 0, [", end = " ")
for i in lf(0):
  print(i,", ", end = " ")
print("... ]")

print("If x = 1, [", end = " ")
for i in lf(1):
  print(i,", ", end = " ")
print("... ]")
```

```python

#Problem 2
def FindIterations(f,x):
  n = 0
  answer = [x]
  while n < 49:
    try:
      x = f(x)
      answer.append(x)
      n += 1
    except:
      break
      print("It's too big")
  return answer



```

```python
#Problem 3
x = sy.symbols("x")
eps = x**2 - 1
f = sy.lambdify(x,eps)

plt.rcParams['figure.figsize'] = [3, 3]

plt.title("x = 1.1")
plt.plot(np.arange(len(FindIterations(f,1.1))),FindIterations(f,1.1), ".")
plt.show()

plt.title("If x = 1.2")
plt.plot(np.arange(len(FindIterations(f,1.2))),FindIterations(f,1.2), ".")
plt.show()

plt.title("If x = 1.3")
plt.plot(np.arange(len(FindIterations(f,1.3))),FindIterations(f,1.3), ".")
plt.show()

plt.title("If x = 1.4")
plt.plot(np.arange(len(FindIterations(f,1.4))),FindIterations(f,1.4), ".")
plt.show()

plt.title("If x = 1.5")
plt.plot(np.arange(len(FindIterations(f,1.5))),FindIterations(f,1.5), ".")
plt.show()

plt.title("If x = 1.6")
plt.plot(np.arange(len(FindIterations(f,1.6))),FindIterations(f,1.6), ".")
plt.show()

plt.title("If x = 1.7")
plt.plot(np.arange(len(FindIterations(f,1.7))),FindIterations(f,1.7), ".")
plt.show()
```

Proble 3;
All elements in each iterate function eventually is bounded by two points, 0 and -1. We got OverflowError: (34, 'Numerical result out of range'). This may be because when we plot x= 1.7, it isn't bounded to 0 and -1 as others do. Since the f(x) goes to infinity, there is an over flow error.

Problem 6 \\
When we solve ${x^2 - 1 = x}$, we got ${x = \frac{\sqrt5+1}{2},\frac{-\sqrt5+1}{2}}$.

```python
x0 = (np.sqrt(5)+1)/2
plt.plot(np.arange(len(FindIterations(f,x0))),FindIterations(f,x0), ".")
plt.show()
```

From the plot above, when ${x = x_0}$ we observe that each iterate function gives the same ${x_0}$; ${f(x_0) = x_0}$. This is called Golden Ratio.
