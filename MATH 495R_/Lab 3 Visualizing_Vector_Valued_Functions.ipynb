For our first example we are going to plot the curve $\mathbf{r}(t) =\langle \sin(t), \cos(t), t\rangle$, which is a parametrized curve in 3-space.  To do this first we create a collection of values for $t$, and then find the corresponding values of $x, y$ and $z$.  As you can see below, plotting in three dimensions is very similar to plotting in two dimensions.

```
#create values for the parameter t, then find the parameterized values for x, y and z
t = np.linspace(-5, 5, 50)
x=np.sin(t)
y=np.cos(t)
z=t
#Create figure and axes, set axes limits.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim3d([-1.2,1.2])
ax.set_ylim3d([-1.2,1.2])
ax.set_zlim3d([-5.2,5.2])

#Plot the curve.
ax.plot(x,y,z)
plt.title("Plot of $< \sin(t), \cos(t), t >$")
plt.show()
```

Of course, not every space curve is so straight forward.  Below you will see an example of a *torroidal spiral* (i.e., a spiral on a torus).  Pay close attention to the values that were adjusted between the two examples.  In the spiral the x and y limits were increased and the z limits were decreased to better accomodate the curve.  Additionally the number of parameter points was changed from 50 to 500 so the curve will appear more smooth. Lasting we introduced the command *ax.view_int* which allows us to change the elevation in which we view the plot.  (You can also change the azimuth as well).

```
#create values for the parameter t, then find the parameterized values for x, y and z
t = np.linspace(0, 2*np.pi, 500)
x=(4+np.sin(20*t))*np.cos(t)
y=(4+np.sin(20*t))*np.sin(t)
z=np.cos(20*t)
#Create figure and axes, set axes limits.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim3d([-5,5])
ax.set_ylim3d([-5,5])
ax.set_zlim3d([-1.2,1.2])

#Plot the curve.
ax.plot(x,y,z)
#this adjusts the viewing angle of the plot
ax.view_init(elev=30)
plt.title("Torroidal Spiral")
plt.show()
```
