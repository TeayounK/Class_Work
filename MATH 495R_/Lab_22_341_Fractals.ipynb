```python
#Problem 1

initializeTurtle()

length = 100
angle = 120
right(30)
for i in range(3):
  forward(length)
  right(angle)
```

```python
#Problem 2
initializeTurtle()
def draw_triangle(side_length):
  angle = 120
  right(30)
  for i in range(3):
    forward(side_length)
    right(angle)
  return left(30)


draw_triangle(100)
```

```python
#Problem 3
initializeTurtle()
size_length = 100

draw_triangle(size_length)

right(30)
forward(size_length)
left(30)
draw_triangle(size_length)

right(150)
forward(size_length)
left(150)
draw_triangle(size_length)
```

```python
#Problem 4
initializeTurtle()
speed(3)

def sierpinski(start_length, min_length):
  if start_length < min_length:
    return
  else:
    start_position = pos()
    draw_triangle(start_length)
    sierpinski(start_length/2,min_length)
    
    right(30)
    penup()
    forward(start_length/2)
    pendown()
    left(30)
    sierpinski(start_length/2,min_length)
    goto(start_position)

    right(90)
    penup()
    forward(start_length/2)
    pendown()
    left(90)
    sierpinski(start_length/2,min_length)
    goto(start_position)
```
