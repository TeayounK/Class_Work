```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Problem 1
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)


X,Y = np.meshgrid(x,y)


u = Y**2 - 2*X*Y
v = 3*X*Y-6*x**2


fig, ax = plt.subplots(figsize=(5,5))


ax.quiver(X,Y,u,v)


ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')


plt.show()
```

```python
#Problem 2
x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)


X,Y = np.meshgrid(x,y)


u = X**3+X*Y**2-2*X*np.sqrt(X**2+Y**2)
v = Y*X**2+Y**3-2*Y*np.sqrt(X**2+Y**2)


fig, ax = plt.subplots(figsize=(5,5))


ax.quiver(X,Y,u,v)


ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')


plt.show()
```

```python
#Problem 3
#a)
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)


X,Y = np.meshgrid(x,y)


z = np.log(1 + X**2 + 2*Y**2)
dy, dx = np.gradient(z)


fig, ax = plt.subplots(figsize=(5,5))


ax.quiver(X,Y,dx,dy)


ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

ax.contour(X,Y,z)


plt.show()
```

```python
#Problem 3
#b)
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)


X,Y = np.meshgrid(x,y)


z = np.cos(X) - 2*np.sin(Y)
dy, dx = np.gradient(z)


fig, ax = plt.subplots(figsize=(5,5))


ax.quiver(X,Y,dx,dy)


ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

ax.contour(X,Y,z)


plt.show()
```

```python
#Problem 4
# a) F(x, y, z) = i + 2j + 3k
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-1, 1, 0.2),
                      np.arange(-1, 1, 0.2),
                      np.arange(-1, 1, 0.8))

# Make the direction data for the arrows
u = 1
v = 2
w = 3

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()

```

```python
#Problem 4
# d) F(x, y, z) = sin yi + (x cos y + cos z)j  y sin zk
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-1, 1, 0.2),
                      np.arange(-1, 1, 0.2),
                      np.arange(-1, 1, 0.8))

# Make the direction data for the arrows
u = np.sin(y)
v = x*np.cos(y)+np.cos(z)
w = y*np.sin(z)

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()
```
