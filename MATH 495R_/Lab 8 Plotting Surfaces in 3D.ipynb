```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Problem 1
 
a = -10 # Interval
b = 10 # Interval

x = np.linspace(a,b,100) 
y = x.copy() 
X, Y = np.meshgrid(x, y) 
Z_a = (X**2 - Y**2) / (X**2 + Y**2) 
Z_b = X*Y**2 - X**3
Z_c = (2*X**2 + 3*X*Y + 4*Y**2) / (3*X**2 + 5*Y**2)


#Plot the heat map off over the 2-D domain. 
plt.subplot(221) 
plt.pcolormesh(X, Y, Z_a, cmap="viridis") 
plt.colorbar() 

plt.subplot(222) 
plt.pcolormesh(X, Y, Z_b, cmap="viridis") 
plt.colorbar() 

plt.subplot(223) 
plt.pcolormesh(X, Y, Z_c, cmap="viridis") 
plt.colorbar() 

plt.xlim(a, b) 
plt.ylim(a, b) 
plt.show()
```

```python
#Problem 2
plt.subplot(221) 
plt.contour(X, Y, Z_a, 15, cmap="coolwarm") 
plt.colorbar() 

plt.subplot(222) 
plt.contour(X, Y, Z_b, 15, cmap="coolwarm") 
plt.colorbar() 

plt.subplot(223) 
plt.contour(X, Y, Z_c, 15, cmap="coolwarm") 
plt.colorbar() 
plt.show()
```

```python
#Problem 3
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
#Plotabasicwireframe. 
ax.plot_wireframe(X, Y, Z_a, rstride=10, cstride=10) 

fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
#Plotabasicwireframe. 
ax.plot_wireframe(X, Y, Z_b, rstride=10, cstride=10) 

fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
#Plotabasicwireframe. 
ax.plot_wireframe(X, Y, Z_c, rstride=10, cstride=10) 


plt.show()
```

```python
#Problem 4
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
#Plotabasicwireframe. 
ax.plot_surface(X, Y, Z_a, rstride=10, cstride=10) 

fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
#Plotabasicwireframe. 
ax.plot_surface(X, Y, Z_b, rstride=10, cstride=10) 

fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
#Plotabasicwireframe. 
ax.plot_surface(X, Y, Z_c, rstride=10, cstride=10) 
plt.show()
```
