```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
 

#Problem 1
n = 100
current = 0
L = 0
target = 0

pos_terms = []
neg_terms = []
pos_ind = 0
neg_ind = 0

rearr = []

x,y = sy.symbols("x,y")
y = (-1)**x/(x+1)
an = sy.lambdify(x,y)

#To get lists of positive and negative terms
for i in range(n+1):
  pos_terms.append(an(2*i))
  neg_terms.append(an(2*i+1))

#Rearrangement partial sum
current = pos_terms[0]
rearr.append(current)
while(pos_ind < len(pos_terms) and neg_ind < len(neg_terms)):
  if current > target:
    current += neg_terms[neg_ind]
    neg_ind += 1
  elif current < target:
    current += pos_terms[pos_ind]
    pos_ind += 1
  rearr.append(current)
  
print("The current is as close as ",current, " to the target at the end of this process.")
```

```python
#Problem 3

def rearrange_sequence(a,target):
  n = len(a)
  end_current_value = 0
  
  pos_terms = []
  neg_terms = []
  pos_ind = 0
  neg_ind = 0

#To get lists of positive and negative terms
  for i in range(n):
    if a[i] >= 0:
      pos_terms.append(a[i])
    else:
      neg_terms.append(a[i])

#Rearrangement partial sum
  end_current_value = pos_terms[0]
  rearranged_end_sequence = []
  rearranged_end_sequence.append(end_current_value)
  while(pos_ind < len(pos_terms) and neg_ind < len(neg_terms)):
    if end_current_value > target:
      end_current_value += neg_terms[neg_ind]
      neg_ind += 1
      rearranged_end_sequence.append(end_current_value)
    elif end_current_value < target:
      end_current_value += pos_terms[pos_ind]
      pos_ind += 1
      rearranged_end_sequence.append(end_current_value)
    else:
      pos_ind += 1
      neg_ind += 1


  return [end_current_value, rearranged_end_sequence]
```

```python
#Problem 4
n = 100

x,y = sy.symbols("x,y")
y = (-1)**x/x*(sy.log(x))
bn = sy.lambdify(x,y)

b =[]

for i in range(2,n+3):
  b.append(bn(i))

print(rearrange_sequence(b,0))
print(rearrange_sequence(b,5))
print(rearrange_sequence(b,-5))
print(rearrange_sequence(b,-0.159869))

#Differences beween the current and the target
print(abs(rearrange_sequence(b,0)[0] - 0))
print(abs(rearrange_sequence(b,5)[0] - 5))
print(abs(rearrange_sequence(b,-5)[0] - (-5)))
print(abs(rearrange_sequence(b,-0.159869)[0] - (-0.159869)))
```
