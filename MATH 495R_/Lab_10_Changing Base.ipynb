```python
import numpy as np
import sympy as sy

# Problem 1
def to_base_b_int(b,n):
  if n <= 0:
    return 0
  coeff = []
  list_of_i = []
  text = str()
  i = 0

  while(n != 0):
    # For decimals
    if (i < 0):
      while (n != 0):
        while (b**i < n):
          i += 1
        if (b**i != n): #and (n >=1):
          i -= 1
          a = n//b**(i)
        else :
          i -= 0
          a = n//b**(i)
        coeff.append(a)
        list_of_i.append(i)
        if ( i < -7):
          n = 0
        else:
          n = n % b**(i)
    #For numbers
    else:
      i = 0
      while (b**i < n):
        i += 1
      if (b**i != n): 
        i -= 1
        a = n//b**(i)
      else :
        i -= 0
        a = n//b**(i)
      coeff.append(a)
      list_of_i.append(i)
      n = n % b**(i)

  
  for i in range(len(coeff)):
    if i != (len(coeff)-1):
      text += ("("+str(coeff[i])+")"+"*"+str(b)+"^"+str(list_of_i[i])+" + ")
    else :
      text += ("("+str(coeff[i])+")"+"*"+str(b)+"^"+str(list_of_i[i]))
  return text



print("Problem 1-a)")
print("Base 5")
print(to_base_b_int(5,123))
print(to_base_b_int(5,495))
print(to_base_b_int(5,31415926))
print("Base 7")
print(to_base_b_int(7,123))
print(to_base_b_int(7,495))
print(to_base_b_int(7,31415926))
print()
print("Problem 1-b)")
print(to_base_b_int(8,(1/2)))
print(to_base_b_int(8,-2))
```
